```python
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound
from IPython.display import FileLink

input_file = "chunk_6.csv"  # your file
df = pd.read_csv(input_file)

if "video_id" not in df.columns:
    raise ValueError("No column named 'video_id' found")

video_ids = []
transcripts = []

ytt = YouTubeTranscriptApi()  # create instance once

for vid in df["video_id"]:
    print("Processing:", vid)
    try:
        fetched = ytt.fetch(vid)
        # combine all snippet texts
        text = " ".join([snippet.text for snippet in fetched])
    except (TranscriptsDisabled, NoTranscriptFound):
        text = "Transcript not available"
    except Exception as e:
        text = f"Error: {e}"
    video_ids.append(vid)
    transcripts.append(text)

out_df = pd.DataFrame({
    "video_id": video_ids,
    "transcript": transcripts
})

output_file = "video_transcripts6.xlsx"
out_df.to_excel(output_file, index=False)
print("Saved:", output_file)

print("\nPreview:")
print(out_df.head())

FileLink(output_file)
```

```python

dfs = []

# Loop from 1 to 6
for i in range(1, 7):
    filename = f"video_transcripts{i}.xlsx"
    if os.path.exists(filename):
        df = pd.read_excel(filename)
        df["source_file"] = filename  # optional: track source file
        dfs.append(df)
    else:
        print(f"Skipped: {filename} not found")

# Merge all dataframes
if dfs:
    merged_df = pd.concat(dfs, ignore_index=True)
    output_file = "video_transcripts_merged.xlsx"
    merged_df.to_excel(output_file, index=False)
    print(f"Merged file saved as '{output_file}'")
else:
    print("No files found to merge.")
```
